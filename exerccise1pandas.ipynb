```python
import pandas as pd

data = {
    "Name": ["Aman", "Saurav", "Karan", "Shubham", "Abhishek", None, "Sameer"],
    "Age": [22, 23, None, 22, 25, 24, 26],
    "Marks": [85, None, 88, 90, 56, 72, None],
    "Subject": ["Math", "Physics", "Math", "Chemistry", "Math", "Physics", "Chemistry"]
}

df = pd.DataFrame(data)

print(df)
```

```python

print(df.shape)
print(df.dtypes)
print(df.describe())
```

```python
print(df[["Name", "Marks"]])
print("---------------")


print(df[df["Marks"] > 80])
print("---------------")


print(df[(df["Age"] > 22) & (df["Marks"] > 80)])
print("---------------")

print(df.groupby("Subject").size())
```

```python
df = pd.DataFrame(data)

print(df.isna().sum())

print(df[df["Marks"].isna()])

cleaned_df = df.dropna()

print(df.shape)
print(cleaned_df.shape)
```
